```python
from src.multiclassifier import OWMultiClassifier
```

```python
import pickle
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Load the pickle file
print("Loading datafile...")
with open(f"open_train_33.pkl", "rb") as f:
    train = pickle.load(f)
    X_train = pd.DataFrame(train["X"])
    X_train.columns = train["feature_names"]
    y_train = pd.DataFrame({
        "y_binary": train["y_binary"],
        "y_true": train["y_true"]
    })

with open(f"open_val_33.pkl", "rb") as f:
    val = pickle.load(f)
    X_val = pd.DataFrame(val["X"])
    X_val.columns = val["feature_names"]
    y_val = pd.DataFrame({
        "y_binary": val["y_binary"],
        "y_true": val["y_true"]
    })

print("Loading complete.")
print("Training data size:\t", len(X_train))
print("Validation data size:\t", len(X_val))
```

```python
from src.train_tuning import Model

owc = Model(True) # open-world binary classifier
# owc.load_model(file_name = "")

cwc = Model(False) # closed-world multi classifier
# cwc.load_model(file_name = "")
```

```python
model = OWMultiClassifier(owc, cwc)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_val, y_val)
```

```python
accuracies = model.k_fold_cv(X_train, y_train)
```

```python
np.mean(accuracies)
```

```python
model.save()
```
