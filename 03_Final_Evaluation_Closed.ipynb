```python
import numpy as np
import pandas as pd
import pickle

from src.train_tuning import Model
from src.evaluation_multi import show_confusion_matrix, per_class_accuracy, overall_metrics, top_k_accuracy
from src.evaluation_final import ClosedWorldPipeline

model_wrapper = Model(False)
```

# Load Data & Model

```python
with open(f"./data/preprocessed/close_test_33.pkl", "rb") as f:
    test = pickle.load(f)
    X_test = pd.DataFrame(test["X"])
    X_test.columns = test["feature_names"]
    y_test = test["y"]

print("Loading complete.")
print("Test data size:\t", len(X_test))
```

```python
multi_classifier = model_wrapper.load_model("./models/multi/stacked")
```

```python
pipeline = ClosedWorldPipeline(multi_classifier=multi_classifier)
```

# Evaluation

```python
y = y_test
y_pred, y_prob = pipeline.predict(X_test)

n_classes = len(np.unique(y))
class_names = np.unique(y)
```

```python
cm = show_confusion_matrix(y, y_pred, class_names)
```

```python
per_class_accuracy(cm, class_names)
```

```python
overall_metrics(y, y_pred, y_prob, n_classes)
```

```python
top_k_accuracy(y, y_prob, class_names)
```
